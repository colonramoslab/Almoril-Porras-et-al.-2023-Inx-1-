```python
import pandas as pd

def get_filtered_in_data(full_csv_path, filtered_csv_path, data_csv_path):
    # Load the two datasets
    full_df = pd.read_csv(full_csv_path, header=None)
    filtered_df = pd.read_csv(filtered_csv_path, header=None)

    # Get the number of columns in both dataframes
    full_columns = set(full_df.columns)
    filtered_columns = set(filtered_df.columns)

    # Find the intersection between the two sets of columns
    filtered_in_indices = list(full_columns & filtered_columns)

    # Load the data CSV
    data_df = pd.read_csv(data_csv_path)

    # Extract the data from columns 3-5 for the filtered in tracks
    filtered_in_data = data_df.iloc[filtered_in_indices, 2:6]
    
    filtered_in_data = calculate_percent_omega(filtered_in_data)
    return filtered_in_data

def calculate_percent_omega(df):
    # Calculate total turns
    df['Total Turns'] = df['Omega Turns'] + df['Blip Turns']
    # Calculate percent omega turns and add it as a new column
    df['Percent Omega Turns'] = df['Omega Turns'] / df['Total Turns'] * 100
    
    df['Total Turns w Sharps'] = df['Omega Turns'] + df['Blip Turns'] + df['# Sharp Turns']
    df['Percent Sharp Turns'] = df['Omega Turns'] / df['Total Turns w Sharps'] * 100
    return df
```
